```python
import pandas as pd

# Create a sample DataFrame
data = {'a': [2, 1, 3, 1],
        'b': [1, 2, 2, 3]}
df2 = pd.DataFrame(data)

# Sort the DataFrame by column 'b' first, then by column 'a'
sorted_df = df2.sort_values(by=["b", "a"])

print(sorted_df)
```
