## Control flow

Control flow is the order that which instructions are executed in a program. A control statement is a statement that determines the control flow of a set of instructions. There are three fundamental forms of control that programming languages provide.

1. Sequential control.
2. Selection control.
3. Iterative control.

- Sequential control is an implicit form of control in which instructions are executed in the
order that they are written. A program consisting of only sequential control is referred to as a
“straight-line program.” While iterative control is provided by an iterative control statement that repeatedly executes instructions. Each is based on a given condition. Collectively a set of instructions and the control statements controlling their execution is called a control structure.

# Python conditionals.

 Conditional statements provide a programmer with the flexibility to determine what actions to take based on a particular scenario happening during runtime.

In Python, blocks are an essential component when it comes to treating a set of statements as a single unit. A block is regarded as a single statement that consists of one or more expressions. Blocks play an essential role in Python programming because they help limit the scope of names and variables, thereby making the code structured and easy to read.

In Python, conditionals are typically implemented using the following statements:

1. if: The if statement allows you to execute a block of code if a condition is true.

2. elif: The elif statement is short for "else if" and is used to check additional conditions if the condition for the if statement is False. It is optional and can appear multiple times within an if statement.

3. else: The else statement is used to define a block of code that should be executed if none of the preceding conditions are true.

Conditionals evaluate a condition or expression that results in a Boolean value (True or False). Based on the evaluation of the condition, the program decides which code block(s) to execute.


## if statement

The simplest conditional statement in Python is the "if statement." An if statement is a decision-making statement that checks whether a certain condition is true or false.

The syntax of such a statement starts with the "if" keyword followed by a condition that is wrapped in parentheses.

The condition can be a boolean expression or an expression that can be evaluated as a boolean expression. If the condition is true, the statements within the block will be executed; otherwise, the program exits the if statement.

```
if condition:
    #code block
```

````python
x = 10

if x > 15:
    print("x is greater than 5")
````

1. We start by defining a variable x and assigning it the value 10.

2. The "if" statement is used to check whether the condition x > 5 is true or false. In this case, it evaluates to True because x is indeed greater than 5.

3. If the condition is true, the indented block of code under the "if" statement is executed. In this case, it prints the message "x is greater than 5."


## Truthy & Falsy Values

In Python, values are considered either "truthy" or "falsy" when used in conditional expressions like if statements. These terms refer to whether a value is treated as equivalent to True or False in a boolean context. Here are some common truthy and falsy values in Python:

## Truthy values (equivalent to True in a boolean context):

- Non-empty sequences: Lists, tuples, and strings with at least one element are considered truthy.
- Non-zero numeric values: Any non-zero integer or float is considered truthy.
Non-empty dictionaries and sets.
- You can also check truthy or falsy values using bool function.

````python
if [1, 2, 3]:
    print("This list is truthy.")

if 42:
    print("The number 42 is truthy.")

if {"key": "value"}:
    print("This dictionary is truthy.")

print(bool("yes"))
````

## Falsy values (equivalent to False in a boolean context):

1. The None value.
2. Numeric zero: The integer 0 and the float 0.0 are considered falsy.
3. Empty sequences: Empty lists, tuples, strings, dictionaries, and sets are all falsy.

You can use the not keyword to negate a condition.

````python
if not None:
    print("This is falsy.")

if not 0:
    print("The number 0 is falsy.")

if not "":
    print("This string is falsy.")

print(bool([]))
````

## If-elif-else statements.

In cases where you need to test multiple conditions, you can use the if-elif-else statement. The "elif" keyword stands for "else if." This statement checks another condition if the first if statement is false. If neither the if statement conditions nor the elif statement conditions are true, then the else block statement executes.

```
if condition1:
    #code block1
elif condition2:
    #code block 2
elif condition3:
    #code block 3
else:
    #code block 4
```


````python
grade = 85

if grade >= 90:
    print("A")
elif grade >= 80:
    print("B")
elif grade >= 70:
    print("C")
elif grade >= 60:
    print("D")
else :
  print("Why not go to IOI")
````

````python
x = 7

if x>=7 or x/0 >5 :
  print("Amazing")
````

1. We define a variable grade and assign it the value 85.

2. The "if-elif" statement checks the conditions one by one, from top to bottom, until it finds the first condition that is true.

3. First, it checks if the grade >= 90, which evaluates to False because 85 is not greater than or equal to 90. So, it moves to the next condition.

4. The second condition grade >= 80 is checked. In this case, it's true because 85 is greater than or equal to 80. So, the code block under this condition is executed, which prints "B."

5. Since the second condition is true, it doesn't check the remaining conditions (i.e., it doesn't check if grade >= 70, grade >= 60  condition). It exits the entire "if-elif" block after executing the code block associated with the first true condition.

In this example, because the grade is 85, it falls into the "B" range, and "B" is printed. The "if-elif" statement is particularly useful when you have multiple conditions to check, and you want to execute the block of code associated with the first true condition, skipping the rest.


## Short-circuit evaluation

Short-circuit evaluation is a feature in Python where logical expressions are evaluated from left to right, and the evaluation stops as soon as the outcome of the entire expression can be determined.

In Python, the logical operators "and" and "or" are used for combining multiple conditions.

```
if condition1 and condition2:
    # Code to execute when both condition1 and condition2 are true

if condition1 or condition2:
    # Code to execute when either condition1 or condition2 is true (or both)\
```

````python
x = 5
y = 10

# The "if" statement checks two conditions using the "and" operator
if x > 0 and y < 15:
    print("Both conditions are true.")
````

1. We initialize two variables, x with a value of 5 and y with a value of 10.

2. In the "if" statement, there are two conditions connected by the "and" operator. Both conditions must be true for the code block under this "if" statement to execute.

3. The first condition checks if x > 0. Since x is indeed greater than 0 (it's 5), this condition is true.

4. The second condition checks if y < 15. y is also less than 15 (it's 10), so this condition is also true.

Since both conditions are true, the code block under this "if" statement is executed, and it prints "Both conditions are true."

Conditional assignment with or.

You can use the or operator for conditional assignment. If the left-hand side is falsy (e.g., None, 0, False), the right-hand side is assigned. This is a concise way to provide default values.



````python
# User's preferred language from a user profile (could be None)
preferred_language = None

# Determine the user's language, with a default if not specified
user_language = preferred_language or "English"

print(f"User's preferred language is: {user_language}")
````

## Nested if else condition

The nested if-else syntax allows you to have an if-else statement inside another if or else block. This allows for more complex branching and decision-making in your code.

```
if condition1:
    # Code block executed when condition1 is true
    if condition2:
        # Code block executed when both condition1 and condition2 are true
    else:
        # Code block executed when condition1 is true and condition2 is false
else:
    # Code block executed when condition1 is false
```


````python
x = 4

if x > 5:
    print("x is greater than 5.")

    if x % 2 == 0:
        print("x is even.")
    else:
        print("x is odd.")
else:
    print("x is not greater than 5.")
````

1. We define a variable x and assign it the value 10.

2. The outermost "if" statement checks whether x is greater than 5. Since 10 is indeed greater than 5, the code block under this "if" statement is executed.

3. Inside the outer "if" block, there's another "if-else" statement, which is nested. This nested "if-else" statement further examines the value of x.

4. The nested "if" statement checks if x is even by evaluating the condition x % 2 == 0. In this case, 10 is divisible by 2 with no remainder, so this condition is true, and it prints "x is even."

5. The "else" block of the nested "if-else" statement is executed if the condition in the nested "if" statement is false. However, since 10 is even, the "else" block is skipped.

6. After completing the nested "if-else" statement, the program continues with the outer "if-else" statement. Since the outer "if" condition was true, we did not execute the "else" block under the outer "if-else" statement.

In nested if-else statements we first check if x is greater than 5 and, if true, proceed to check whether x is even or odd using a nested "if-else" statement. In this specific case, because x is greater than 5 and even, it prints "x is greater than 5" and "x is even."


## Ternary operators:

Ternary operators are a shorthand way of writing if-else statements in Python. They are useful when you want to write simple conditional statements.


````python
num = 5
result = "Positive" if num >= 0 else "Negative"
print(result)
````

In this example, if the number is greater than or equal to 0, the result variable is set to "Positive". Otherwise, it is set to "Negative". The ternary operator is written in one line, which makes the code more concise.

## Match Case:

In Python, the match statement was introduced in PEP 634 (Python Enhancement Proposal) and added in Python 3.10. It provides a way to perform pattern matching, allowing you to match a value against multiple patterns and execute code based on the first matching pattern. It's similar to the switch statement in other programming languages.

```
match expression:
    case pattern1:
        # Code block to execute if the expression matches pattern1
        # ...
    case pattern2:
        # Code block to execute if the expression matches pattern2
        # ...
    case pattern3:
        # Code block to execute if the expression matches pattern3
        # ...
    ...
    case _:
        # Code block to execute if no patterns match (optional)
        # ...
```


````python
# Define a variable to represent a day of the week
day_number = 5

# Using the match statement to determine the day of the week
match day_number:
    case 1:
        print("Today is Monday")
    case 2:
        print("Today is Tuesday")
    case 3:
        print("Today is Wednesday")
    case 4:
        print("Today is Thursday")
    case 5:
        print("Today is Friday")
    case 6:
        print("Today is Saturday")
    case 7:
        print("Today is Sunday")
    case _:
        print("Invalid day number")

````

1. We define a variable day_number and assign it the value 5, representing Friday.

2. The match statement is used to match the value of day_number against different cases. It starts with the match keyword followed by the value to be matched (in this case, day_number).

3. Each case block represents a possible value of day_number. When a case matches the value, the corresponding block of code is executed.

4. In this example, the value of day_number is 5, which matches the case 5:. Therefore, the code block under case 5: is executed, printing "Today is Friday."

5. If none of the cases match, the _ case serves as a catch-all. If day_number were a value other than 1 through 7, it would execute the code block under case _:, which prints "Invalid day number."

The match statement allows for concise and readable code when you need to perform pattern matching against a value to determine the appropriate action to take.
