**How does python handle unlimited size of int value ?**

When we say Python can store any number of any size in an integer variable, it's actually saying that Python can handle integers of virtually unlimited size, limited only by the available memory in your computer.

To understand this better let's have a look at some information connected to Python's virtually unlimited size.

1. Memory: The ability to handle unlimited integers is constrained by the amount of available memory on your computer. If your integer becomes too large to fit in your computer's memory, you will run into memory-related issues.

2. Performance: Operating on very large integers can be slower than operating on smaller ones because of the increased memory allocation and computation required.

3. Precision: Python's unlimited integers have arbitrary precision, meaning they can represent numbers with as many digits as you have memory for, but this precision comes at the cost of increased memory usage and potentially slower computations.

Considering all of the above factors we can conclude that the size of integer is unlimited with Python's perspective but has some limitations in the hardware perspective.

Why does indexing starts from zero?

This is often confusing for new developersâ€Šâ€”â€Šwhy do we start numeration from zero? The answer to this lies in the design of the language itself. What happens in many programming languages including Python is data is stored in memory as a contiguous block of bytes.

To understand this you must be aware of offset first. Offset refers to a numeric value that represents the position or location of a particular element or piece of data refering to some reference point within the memory

When you access a particular element of a data, you are essentially specifying an offset from the beginning of that block. Starting indexing from 0 aligns with the idea that the first element is at the beginning of the block (i.e., 0 offset), making memory addressing more straightforward.

This is why indexing starts from zero in Python.

It's important to note that not all languages follow this convention. Some languages, like Fortran, use one-based indexing, which starts counting from 1. Others, like Lua, allow you to choose the starting index based on your preferences. The choice of indexing convention often depends on the language's design goals and historical influences.

**String Manipulation.**

In the world of Python strings, there's a whole universe of possibilities for manipulating text. From slicing and changing case to replacing substrings, let's explore some of the key techniques that will empower you to master the power of text manipulation.

**String Slicing.**

With string slicing in Python, you can select specific portions of a string by specifying the starting and ending indices.

The indexing starts at 0, allowing you to easily retrieve individual characters or even whole sections.


`Note: Remember that slicing doesn't modify the original string; it creates a new string containing the specified portion of the original string`

The syntax for slicing is string [start:end], where start is the index of the first character to include in the slice (inclusive), and end is the index of the first character to exclude from the slice (exclusive).

```python
bio = "I am Mithun"

# Slice from the index 5 (inclusive) to index 11 (exclusive)
slice1 = bio[5:11]
print(slice1)  # Output: 'Mithun'
```

**Slicing from the start**

When slicing a string in Python, you can omit the start index to slice from the beginning of the string. If you don't specify the start index, Python assumes that you want to start from the first character (index 0). This is known as slicing from the start.

```python
bio = "I am Mithun"

# Slice from the beginning to index 5 (exclusive)
slice1 = bio[:5]
print(slice1)  # Output: 'I am'

# Slice from the beginning to index 7 (exclusive)
slice2 = bio[:7]
print(slice2)  # Output: 'I am Mi'
```

**Slicing till the end**

When slicing a string in Python, you can omit the end index to slice until the end of the string. If you don't specify the end index, Python assumes that you want to include all characters until the end. This is known as slicing till the end.

```python
bio = "I am Mithun"

# Slice from index 5 to the end of the string
slice1 = bio[5:]
print(slice1)  # Output: 'Mithun'
```

```python
print(bio[-6:-4])
```

**Modifying Strings**

In Python, strings are immutable. Since you cannot change individual characters in a string directly, you can create new strings by combining or slicing the original string, and then assign the modified string to a new variable.

**upper()**

upper() is a string method used to convert all the characters in a string to uppercase. It returns a new string with all uppercase letters while leaving any non-alphabetic characters unchanged.

```python
bio = "I am Mithun"

# Convert the entire string to uppercase
uppercase_bio = bio.upper()

print(uppercase_bio)  # Output: 'I AM MITHUN'
```

**lower()**

The lower() is a string method used to convert all the characters in a string to lowercase. Just like the upper() method, the lower() method does not modify the original string (text) but instead creates a new string with all lowercase characters. This behavior is due to the immutability of strings in Python.


```python
bio = "I am Mithun"

# Convert the entire string to lowercase
lowercase_bio = bio.lower()

print(lowercase_bio)  # Output: 'i am mithun'
```

**strip()**

strip() is a string method used to remove leading and trailing whitespace characters from a string. By default, it removes spaces, tabs, and newline characters from the beginning and end of the string. It returns a new string with the leading and trailing whitespace removed.


```python
bio = "   I am   Mithun!!   "

# Remove leading and trailing whitespace
stripped_bio = bio.strip()

print(stripped_bio)  # Output: "I am   Mithun!!"
```

**capitalize()**

capitalize() is a string method used to capitalize the first character of a string. It converts the first character to uppercase and leaves the rest of the characters unchanged. If the first character is already in uppercase, the method does not alter it.


```python
name = 'mithun'

print(name.capitalize())  # Output: 'Mithun'

name2 = 'Sanjay'

print(name2.capitalize())  # Output: 'Sanjay'
```

**split()**

split() is a string method used to split a string into a list of substrings based on a specified delimiter. It allows you to break a string into smaller parts whenever the delimiter is encountered. The split() method returns a list of these substrings.

Syntax:

`string.split(sep, maxsplit)`


- sep (optional): The delimiter that specifies where the string should be split. If not provided, the default delimiter is whitespace (spaces, tabs, and newlines).
- maxsplit (optional): Specifies the maximum number of splits to be performed. If not provided, there is no limit to the number of splits.


```python
text = "Hello, How are you?"

# Split the string using the default delimiter (whitespace)
words = text.split()

print(words)  # Output: ['Hello,', 'How', 'are', 'you?']
```

```python
text = "Hello, How are you?"

# Split the string using a custom delimiter (",")
parts = text.split(",")

print(parts)  # Output: ['Hello', ' How are you?']
```

**replace()**

replace() is a string method used to replace occurrences of a substring with a new substring within a string. It returns a new string with the replacements made, leaving the original string unchanged.

Syntax:

`string.replace(old_substring, new_substring, count)`


- old_substring: The substring you want to replace in the original string.
- new_substring: The new substring that will replace occurrences of the old_substring.
- count (optional): Specifies the maximum number of occurrences to replace. If not specified, all occurrences will be replaced.


```python
text = "Hello, Mithun!"

# Replace "Mithun" with "All"
new_text = text.replace("Mithun", "All")

print(new_text)  # Output: "Hi, All!"
```

**Accessing Items in a List:**

You can access individual items in a Python list by using their index number. List indices start from zero, hence the first item in the list has an index of 0 and the last item has an index of n-1 where n is the length of the list.
```python
students = ['Mithun', 'Charan', 'Ranjan']

print(students[-3:-1])  # Output: ['Mithun', 'Charan']
```

### **Adding items to a list**

To add an item to the list, we have 3 methods.

1. append(): To add an item to the end of the list, we use the append() method.

2. insert(): To insert a list item at a specified index, we use the insert() method.

3. extend(): To append elements from another list to the current list, use the extend() method.


```python
students = ['Mithun', 'Charan', 'Ranjan']

students.append('Trilok')

print(students)  # Output: ['Mithun', 'Charan', 'Ranjan', 'Trilok']
```

```python
students = ['Mithun', 'Charan', 'Ranjan']

students.insert(1, 'Trilok')

print(students)  # Output: ['Mithun', 'Trilok', 'Charan', 'Ranjan']
```

```python
students1 = ['Mithun', 'Charan']

students2 = ['Ranjan', 'Trilok']

students1.extend(students2)

print(students1)  # Output: ['Mithun', 'Charan', 'Ranjan', 'Trilok']
```

### **Removing elements from a list**

There are 2 methods to remove an item from the list.

1. remove(): The remove() method removes the specified item.
2. pop(): The pop() method removes the specified index. If you do not specify the index, the pop() method removes the last item.


```python
students = ['Mithun', 'Charan', 'Ranjan', 'Trilok']

students.remove('Trilok')

print(students)  # Output: ['Mithun', 'Charan', 'Ranjan']
```

```python
students = ['Mithun', 'Charan', 'Ranjan', 'Trilok']

students.pop(2)

print(students)  # Output: ['Mithun', 'Charan', 'Trilok']

students.pop()

print(students)  # Output: ['Mithun', 'Charan']
```

## **Tuples.**

Tuples are sequential collections of objects, similar to lists. However, unlike lists, tuples are immutable, which means they cannot be modified after creation. Tuples are created by enclosing comma-separated values in parentheses.

The immutability of tuples is a key characteristic that sets them apart from other data structures in Python. Once a tuple is created, its values cannot be changed or updated. This immutability makes tuples useful when you want to store a collection of values that should not be modified.

**Features of tuples.**

1. Ordered: Tuples keep your items in the order you add them, just like lists.

2. Immutable: Unlike lists, once you create a tuple, you can't change, add, or remove items from it.

3. Heterogeneous: Tuples can hold different types of data, similar to lists.

4. Indexing: You can access items in a tuple by their position, using numbers (indexes), just like you do with lists.

5. Iterable: Tuples can be used in loops to go through each item one by one, just like lists.

6. Size-Fixed: Once you create a tuple, its size doesn't change. It's like a set list of items you can't modify.


### **Creation of Tuples.**

```python
fruits = ('apple', 'mango', 'apple', 'pineapple', 'watermelon')

print(fruits)
```

### **Tuple Methods**

Now that we understand the basics of tuples, let's explore some useful methods that can be applied to tuples.


**count() Method**

The count() method allows you to count the number of occurrences of a specific value within a tuple.


```python
fruits = ('apple', 'mango', 'apple', 'pineapple', 'watermelon')

print(fruits.count('apple'))
```

**index() Method**

The index() method returns the index of the first occurrence of a value in a tuple. If the value is not found, it raises a ValueError.


```python
fruits = ('apple', 'mango', 'apple', 'pineapple', 'watermelon')

print(fruits.index('apple'))  # Output: 0
```

```python
fruits = ('apple', 'mango', 'apple', 'pineapple', 'watermelon')
print(fruits.index('greenapple'))  # Output: ValueError: tuple.index(x): x not in tuple
```

**len() Function**

The len() function is not specific to tuples, but it is commonly used to determine the length of a tuple. It returns the number of elements in the tuple.


```python
fruits = ('apple', 'mango', 'apple', 'pineapple', 'watermelon')

print(len(fruits))  # Output: 5
```

### **Accessing items of a tuple**

There are 4 different ways of accessing items in a tuple.

1. Positive Index: You can access tuple items by referring to the index number(0 to length of tuple-1), inside square brackets.


```python
fruits = ('apple', 'mango', 'apple', 'pineapple', 'watermelon')

print(fruits[1])  # Output: 'mango'
```

2. Negative Indexing: Negative indexing means starting from the end. -1 refers to the last item, -2 refers to the second last item, etc.


```python
fruits = ('apple', 'mango', 'apple', 'pineapple', 'watermelon')

print(fruits[-2])  # Output: 'pineapple'
```

3. Range of index: You can specify a range of indexes by specifying where to start and where to end the range. When specifying a range, the return value will be a new tuple with the specified items. Herein, the start of the range is inclusive and the end of the range is exclusive, i.e., elements are considered from the start index till the end index-1.

```python
fruits = ('apple', 'mango', 'apple', 'pineapple', 'watermelon')

print(fruits[1: 3])  # Output: ('mango', 'apple')
print(fruits[-3: -1])  # Output: '('apple', 'pineapple')'
```

## **Dictionary**

Python dictionaries are powerful data structures that allow you to store and retrieve data using key-value pairs. They are like real-life dictionaries, where you can look up a word and find its meaning. In Python, you can look up a key and get its corresponding value in an instant.

Dictionaries are enclosed in curly braces {} and consist of key: value pairs, separated by commas. The keys in a dictionary must be unique, but the values can be duplicated.


**Features of Dictionary.**

1. Key-Value Pairs: Dictionaries are collections of key-value pairs. Think of them like a phone book, where names (keys) are associated with phone numbers (values).

2. Unordered: Dictionaries don't have a specific order for their elements.

3. Mutable: You can change, add, or remove key-value pairs in a dictionary after creating it.

4. Unique Keys: Keys in a dictionary are unique. You can't have two entries with the same key.

5. Heterogeneous Values: Values in a dictionary can be of different data types. You can store numbers, text, lists, or even other dictionaries as values.


**Creating a Dictionary:**

To create a dictionary, simply enclose key-value pairs in curly braces {}.


```python
student_details = {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru'}

print(student_details)
# Output: {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru'}
```

It is also possible to use the `dict()` constructor to make a dictionary.


```python
student_details = dict(name='Mithun', age=21, city='Bengaluru')

print(student_details)

# Output: {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru'}
```

**Accessing Dictionary Values:**

You can access the values of a dictionary using the corresponding keys.


```python
student_details = {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru'}

print(student_details['name'])  # Output: 'Mithun'
print(student_details['age'])  # Output: 21
print(student_details['city'])  # Output: 'Bengaluru'
```

**Modifying Dictionary Entries:**

Dictionaries are mutable, which means you can change their values or add new key-value pairs.


```python
student_details = {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru'}

student_details['city'] = 'Delhi'

print(student_details)
# Output: {'name': 'Mithun', 'age': 21, 'city': 'Delhi'}
```

### **Dictionary Methods:**

Python provides several built-in methods to perform various operations on dictionaries.


**keys() method.**

The keys() method will return a list of all the keys in the dictionary.


```python
student_details = {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru'}

print(student_details.keys())
# Output: dict_keys(['name', 'age', 'city'])
```

**values() method.**

The values() method will return a list of all the values in the dictionary.


```python
student_details = {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru'}

print(student_details.values()) # Output: dict_values(['Mithun', 21, 'Bengaluru'])
```

**items() method.**

The items() method will return a list of all the entries in the dictionary.


```python
student_details = {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru'}

print(student_details.items())
# Output: dict_items([('name', 'Mithun'), ('age', 21), ('city', 'Bengaluru')])
```

**Adding an item to a dictionary**

Adding an item to the dictionary is done by using a new index key and assigning a value to it.


```python
student_details = {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru'}

student_details['country'] = 'India'

print(student_details)
# Output: {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru', 'country': 'India'}
```

Any other dictionary or any iterable having key-value pairs can also be added to a dictionary using update() method.


```python
student_details = {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru'}

marks_details = {'English': 100, 'Maths': 99, 'Science': '98'}

student_details.update(marks_details)

print(student_details)
# Output: {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru', 'English': 100, 'Maths': 99, 'Science': '98'}
```

### **Removing item from a dictionary**

There are several methods to remove items from a dictionary.

1. The pop() method removes the item with the specified key name.


```python
student_details = {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru'}

student_details.pop('city')

print(student_details)
# Output: {'name': 'Mithun', 'age': 21}
```

2. The popitem() method removes the last inserted item.

```python
student_details = {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru'}

student_details.popitem()

print(student_details)
# Output: {'name': 'Mithun', 'age': 21}
```

3. del: The del keyword removes the item with the specified key name.

```python
student_details = {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru'}

del student_details['age']

print(student_details)
# Output: {'name': 'Mithun', 'city': 'Bengaluru'}
```

4. The clear() method empties the dictionary.

```python
student_details = {'name': 'Mithun', 'age': 21, 'city': 'Bengaluru'}

student_details.clear()

print(student_details)  # Output: {}
```

## **Sets.**

 A set is an unordered collection of unique elements, where each element is unique and immutable.

 This means that a set cannot contain duplicate elements, and the elements themselves cannot be changed once they are added to the set.


**Features of Sets.**

1. Unordered Collection: Sets are collections of items, similar to lists, but they don't have a specific order.

2. Unique Elements: Sets only allow unique elements. If you try to add the same item twice, it will still appear only once in the set.

3. Mutable: You can add or remove items from a set after creating it.

4. Heterogeneous: Sets can hold elements of different data types.

5. No Duplicate Elements: Since sets allow only unique elements, they are excellent for removing duplicates from a list or checking for common elements between collections.


**Creating Sets.**

Creating a set is easy! To create a set, you can use the set() function or the curly braces {}.


```python
my_set = {1, 2, 3}

print(my_set)  # Output: {1, 2, 3}
```

You can also use the set() constructor to create a set.

```python
my_set = set((1, 2,3))

print(my_set)  # Output: {1,2,3}
```

**Adding and Removing Elements.**

Now that we have created our sets, let's learn how to add and remove elements. To add an element to a set, you can use the add() method:


```python
my_set = set()

my_set.add(5)
my_set.add(4)
my_set.add(3)

print(my_set)  # Output: {3, 4, 5}
```

To remove an element from a set, you can use the discard() or remove() method:

1. The discard method will remove the element from the set. If the element does not exist, discard() will not raise an error.


```python
my_set = {3, 4, 5}

my_set.discard(5)

print(my_set)  # Output: {3, 4}

my_set.discard(10)

print(my_set)  # Output: {3, 4}
```

2. The remove method will also remove the element from the set but if the element does not exist, remove() will raise a key error.


```python
my_set = {3, 4, 5}

my_set.remove(5)

print(my_set)  # Output: {3, 4}
```

```python
my_set = {3, 4, 5}

my_set.remove(10)

print(my_set)  # Output: KeyError: 10
```

## **frozenset in Python**

A frozenset in Python is an immutable set, meaning its elements cannot be modified after it is created.

 You can create a frozenset using the frozenset() constructor or by simply enclosing a sequence of elements within frozenset() curly braces.



```python
fs = frozenset([1, 2, 3])
fs1 = frozenset({1, 2, 3})

print(fs)
print(fs1)
```

Frozensets are handy when you need to use a set as a key in a dictionary, store a set in another set, or ensure that a set's contents do not change accidentally.

```python
my_dict = {}
my_set = frozenset([1, 2, 3])
my_dict[my_set] = "Value associated with the set"

print(my_dict)
```

## **Type in Python.**

The type() function in Python is used to determine the type of a variable or value. It returns the type as an object.

For example, if we have a variable x and we want to know its type, we can use the type() function like this: type(x).

```python
name = 'Mithun'
print(type(name))  # Output: <class 'str'>
```

The type() function is a built-in function in Python and can be used with any object or variable. It gives us information about the type of the object or variable, which can be helpful in performing type casting or type checking.

## **Implicit Type Casting.**

Implicit Type Casting is a feature of Python that allows the language to automatically convert variables from one data type to another, without the need for explicit conversion.

In simpler terms, implicit type casting occurs when Python automatically converts one data type to another, based on the context of the operation being performed.

Let's take a look at an example to make things more clear. Suppose we have two variables, x, and y, where x is an integer and y is a float. Now, if we try to add x and y together, Python will automatically convert x to a float before performing the addition. This is called implicit type casting.


```python
x = 5
y = 6.5

print(x+y)  # Output: 11.5
```

## **Explicit Type Casting.**

Explicit type casting, also known as type conversion, is the process of manually converting one data type to another. In Python, explicit type casting can be done using built-in conversion functions such as int(), float(), and str().

Explicit type casting allows you to convert a variable of one type into another type. It is done by specifying the desired type in front of the variable that needs to be converted. This ensures that the variable is treated as the specified type and can be used accordingly.

Let's say you have a variable x with a value of 5, and you want to convert it to a float type. You can use the float() function to explicitly cast the variable to a float type as follows:


```python
x = 5
float_x = float(x)

print(x)  # Output: 5
print(float_x)  # Output: 5.0
```

```python
y = '10'
int_y = int(y)

print(y, type(y))  # Output: 10 <class 'str'>
print(int_y, type(int_y))  # Output: 10 <class 'int'>
```

## **Type Conversion Functions:**

 1. `str()` Function:

- The `str()` function is used to convert a variable into a string data type. This function takes the variable as input and returns its string representation. It is particularly useful when we want to concatenate variables of different data types as strings.

 2. `int()` Function:

- The `int()` function is used to convert a variable into an integer data type. It takes a string or a number as input and returns the corresponding integer value.

- If the input cannot be converted into an integer, it raises a ValueError. It is commonly used when we want to perform mathematical operations or comparisons on variables that are originally in string or float format.

3. `float()` Function:

- The `float()` function is used to convert a variable into a floating-point data type. It takes a string or an integer as input and returns the corresponding floating-point value.

- Similar to the `int()` function, if the input cannot be converted into a float, it raises a ValueError. This function is useful when we want to perform operations that require decimal precision on variables that are originally in string or integer format.
