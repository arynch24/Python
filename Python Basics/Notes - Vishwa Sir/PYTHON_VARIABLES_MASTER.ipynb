**How are variables actually stored in Python?**

Let's now have a look at the behind the scenes of the code.

In Python, the way values are stored in memory is a bit different from low-level languages like C or C++. Behind the scenes, Python uses a system of objects and references. Before actually understanding how values are stored let's have a look at objects and references.

1. Objects:

- In Python, everything is treated as an object. This includes numbers, strings, lists, functions, and even custom objects you create.
- Each object has two parts: the actual data, which stores the value, and additional information about the object (like its type).

2. References:

- Instead of directly storing the objects, Python stores references to objects in memory. A reference is like a pointer or a link to the actual object.

When you create a variable and assign a value to it, you're essentially creating a reference to the object that holds the value.

For example, when you do a = 50, you're creating a reference **a** that points to an integer object containing the value 50.

1_n0MVfelkQrJ_zBym4b3CSA.png

Python uses a memory manager to keep track of all objects and their references. It also handles memory allocation and deallocation automatically. When there are no references to an object (i.e., nothing is pointing to it), Python's memory manager deallocates the memory occupied by that object.

````python
x = 42  # Creates an integer object with the value 42 and a reference x to it.

x = 100  # Creates a new integer object with the value 100 and updates the reference x to point to it.

# Memory manager eventually deallocates the old integer object as it has no references.
````

This approach allows Python to be more flexible. Understanding this object-reference model is essential for efficient and bug-free Python programming, especially when working with large datasets or complex data structures.

**Variable Name Casing Styles**

In Python, there are several casing styles commonly used for variable names and identifiers. Each casing style serves a specific purpose and follows its own conventions. Here are the most common casing styles in Python:

1. snake_case:

- In snake_case, words are separated by underscores (_) in lowercase letters. This style is commonly used for variable names, function names, and module names in Python.


````python
my_variable_name
````

2. camelCase:

- In CamelCase (or CamelCaseWithInitialCapital), words are concatenated without spaces, and each word starts with an uppercase letter. This style is often used for class names and, in some cases, method names.


````python
myVariableName
````

3. PascalCase:

- PascalCase is similar to CamelCase, but it starts with an uppercase letter. This style is typically used for class names in Python.


````python
MyClassName
````

5. _single_leading_underscore:

- Variables starting with a single underscore are considered "weakly private," indicating that they are intended for internal use within a module or class. However, this is just a naming convention and does not enforce privacy.


````python
_private_variable
````

6. __double_leading_underscore:

- Variables starting with a double underscore are used to create "private" attributes, although they are still accessible if needed.


````python
__private_attribute
````

These casing styles are important for maintaining code readability and consistency. Choosing the appropriate style depends on the context and the conventions followed in the Python community or within a specific codebase.

**Python Indentation**

Python is a popular programming language known for its readability and simplicity. One of the key features that sets Python apart from other languages is its use of indentation to define the structure of the code. Indentation refers to the spaces or tabs before a line of code, which determines the hierarchy and grouping of statements.

In Python, indentation is used to indicate blocks of code. It is a way of visually organizing code and making it more readable. Instead of using brackets or keywords like "begin" and "end" to denote code blocks, Python uses indentation. This means that the indentation level of each line determines which code block it belongs to.

Proper indentation is crucial in Python because it directly affects the execution of the code. Since Python relies on indentation, even a small mistake in indentation can lead to syntax errors or unexpected behavior. Therefore, it is essential to maintain consistent and correct indentation to ensure that the code functions as intended.

**Input and Output in Python.**

Input and output (I/O) are essential aspects of any programming language, because they enable communication between a computer program and the user or the external world.

**Getting User Input**

Input is a crucial part of any program. It allows us to interact with the user and gather information for further processing. In Python, we can use the

```
input()
```

 function to prompt the user for input.

 It is important to note that the input provided by the user is then returned as a string.

````python
name = input()
````

You can also prompt a message while accepting user input.

````python
name = input("What's your name? ")
````

Here, the message "What's your name? " appears on the console, and the user can enter their name. The input is stored in the variable `name` as a string for further use.

**Displaying Output**

Once we have collected the necessary input, we often need to display some output to the user.

In Python, we can use the `print()` function to achieve this.

The `print()` function takes one or more arguments and displays them on the console.

Here are some examples of how to use the print() function:

1. Printing Text.

````python
print("Hello All !!")
````

This line of code will display the text “Hello All !!” on the console.

2. Printing Variables.

````python
name = "Mithun"
age = 21
print("My name is", name, "and I am", age, "years old.")
````

The above code prints the text “My name is Mithun and I am 21 years old.” by collecting the values from the variables “name” and “age”.

3. Printing Expressions.

````python
x = 10
y = 5
print("The sum of", x, "and", y, "is", x + y)
````

4. Printing Formatted Strings.

Starting from Python 3.6, Python offers  f-strings, which are a concise and readable way to format strings.

````python
name = "Mithun"
age = 21
print(f"My name is {name} and I am {age} years old.")
````

Python also supports the .format() method for string formatting.

````python
name = "Mithun"
age = 21
print("My name is {} and I am {} years old.".format(name, age))
````

5. Printing with Separator.

````python
x = 1
y = 2
z = 3
print(x, y, z, sep=",")
````

The above code prints the value of variables “x”, “y” and “z” separated by “,” which is specified in the code itself.
